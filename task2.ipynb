# New Section

```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

df.head()
```

# New Section

```python
city_count = df['City'].value_counts()

top_city = city_count.idxmax()

top_city
```

```python
avg_rating = df.groupby('City')['Aggregate rating'].mean()

avg_rating.round(2)
```

```python
top_rating_city = avg_rating.idxmax()

top_rating_city
```

```python
print("City with highest restaurants:", top_city)

print("\nAverage rating per city:\n", avg_rating.round(2))

print("\nCity with highest average rating:", top_rating_city)
```
