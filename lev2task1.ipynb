```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")

df.head()
```

```python
df['Aggregate rating'].describe()
```

```python
df['Aggregate rating'].plot(kind='hist', bins=10)

plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
rating_counts = df['Aggregate rating'].value_counts().sort_index()

rating_counts.head()
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']

df['rating_range'] = pd.cut(df['Aggregate rating'], bins=bins, labels=labels, right=False)

range_counts = df['rating_range'].value_counts().sort_index()
most_common_range = range_counts.idxmax()

print(f"Rating distribution by range:\n{range_counts}")
print(f"The most common rating range is: {most_common_range}")
```
