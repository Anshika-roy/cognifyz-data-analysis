```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")

df.head()
```

```python
df['Has Online delivery'].value_counts()
```

```python
delivery_count = df['Has Online delivery'].value_counts()

percentage = (delivery_count / len(df)) * 100

percentage.round(2)
```

```python
avg_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()

avg_rating.round(2)
```

```python
print("Online Delivery Percentage:\n", percentage.round(2))

print("\nAverage Ratings:\n", avg_rating.round(2))
```
