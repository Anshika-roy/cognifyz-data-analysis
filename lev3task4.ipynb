```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

df.head()
```

```python
chain_counts = df['Restaurant Name'].value_counts()

chains = chain_counts[chain_counts > 1]

chains.head(10)
```

```python
top_chains = chains.sort_values(ascending=False).head(10)

top_chains
```

```python
chain_ratings = df[df['Restaurant Name'].isin(top_chains.index)]

avg_chain_rating = chain_ratings.groupby('Restaurant Name')['Aggregate rating'].mean()

avg_chain_rating.sort_values(ascending=False)
```

```python
chain_votes = chain_ratings.groupby('Restaurant Name')['Votes'].mean()

chain_votes.sort_values(ascending=False)
```

```python
print("Top Restaurant Chains:\n", top_chains)

print("\nAverage Ratings of Chains:\n", avg_chain_rating.round(2))

print("\nAverage Votes (Popularity):\n", chain_votes.round(2))
```
